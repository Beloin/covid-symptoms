```python
import pandas as pd
base = pd.read_csv('./Cleaned-Data.csv')
pd.set_option('display.max_columns', None)
```

```python
base.head()
```

```python
severity_columns_to_delete = ['Severity_Severe', 'Severity_Moderate', 'Severity_Mild']
rename = 'Severity_None'
cleaned_base = base.rename({rename:'None_Severity'}, axis=1).drop([*severity_columns_to_delete, 'Country'], axis=1)
cleaned_base.head()
```

```python
cleaned_base = cleaned_base.drop(['Contact_Dont-Know', 'Gender_Transgender', 'None_Experiencing', 'None_Sympton'], axis=1)
```

```python
cleaned_base.head()
```

```python
# Start Modeling
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

X = cleaned_base.drop('None_Severity', axis=1)
y = cleaned_base['None_Severity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
```

```python
def test_model(model, x, y, x_test, y_test):
    model.fit(x, y)
    pred = model.predict(x_test)
    acc = accuracy_score(y_test, pred)
    cf = confusion_matrix(y_test, pred)
    return acc, cf
```

```python
from sklearn.tree import DecisionTreeClassifier

test_model(DecisionTreeClassifier(max_depth=50), X_train, y_train, X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussian_model = GaussianNB()

test_model(gaussian_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_model = RandomForestClassifier()
test_model(random_model,  X_train, y_train, X_test, y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

input_dim = len(X.columns)
opt = Adam(learning_rate=0.03)


ann_model = Sequential()
ann_model.add(Dense(10, input_dim=input_dim, activation='relu'))
ann_model.add(Dense(12, activation='relu'))
ann_model.add(Dense(5, activation='softmax'))
ann_model.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
ann_model.fit(X_train, y_train, epochs=120, batch_size=20)
```
